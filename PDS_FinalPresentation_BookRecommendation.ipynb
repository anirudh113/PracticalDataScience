<h2>Introduction</h2>

Recommendation models are in heavy demand these days. A good recommendation is a win-win situation for both user and the organization. It helps the user in browsing the relevant items which he/she may be looking for, and at the same time benefits the organization by targeting specific items to users. 

While there are several possible algorithms for recommendation engines, most of them fit into three categories: 

User-User recommendation

Item-Item approaches

Matrix Factorization

In this project have designed an item based recommendation engine. The motivation behind this model is that the item based recommendation model is the most scalable one. Models such as user based and collaborative filtering require computation which may be expensive to do in real time due to large base of users and items. Giving recommendations in such scenario can be very time consuming. Therefore item based recommendation models are used to give fast and relevant recommendations.


The key to item-to-item collaborative filtering’s scalability and performance is that it creates the cosine similarity matrix offline. Once this matrix is calculated, the online component uses this matrix to look up similar items for the user’s selected items in the cart. Thus, it scales independently of the catalog size or the total number of customers. The algorithm is fast even for extremely large data sets. The quality of recommendations is excellent as the algorithm recommends highly correlated items, that are inline with the users taste. Unlike traditional collaborative filtering, the algorithm also performs well with limited user data, and produces good recommendations based on as few as two or three items.


To demonstrate our algorithm, we prompt users to select one or more books from a list of 10 books chosen randomly from our dataset. On selecting any of these books, we recommend three similar books that the user may be interested to read.

## Data Collection

We have scraped data of users and the books they have read and rated from goodreads.com. The phases that we followed for scraping are as follows:

1) In the first phase, we scrape user data. We have chosen the first 100 users of Goodreads for this purpose. Within the scraper we navigate to each of these users' pages and get the ratings of books that they have read. This is written in the bookFile.csv file. There are 5783 distinct books read and rated by these users along with their URLs.

2) The next phase starts with scraping the data of each of the books read by these users, using the book URL. The data from this is written onto the bookFeatures.csv, which consists of features of each of these books, like author, book name, rating statistics and genres.

The scraper works in these two phases and takes around takes around 15 minutes for the first phase and around 400 minutes for the second phase. It is to be noted that we have kept a wait time of 1 sec per request to ensure we are not blocked out.

## Scraper (Users and Books)

### Scraper for User Ratings

```python
users1=['Googlebot/2.1','Mozilla/5.0']
users=[]
book_urls=[]
ratings=[]
for i in range(1,100):

    url = "https://www.goodreads.com/review/list/"+str(i)+"?sort=rating&view=reviews&per_page=30"
    while url is not None:

        agent=random.choice(users1)
        req = Request(url, headers={'User-Agent': agent})
        webpage = urlopen(req).read().decode('utf-8')

        soup = BeautifulSoup(webpage) 

        count=0
        for tr in soup.find_all("tr", class_="bookalike review"):
            row =tr.find("td",class_="field title")
            book_urls.append(row.find("a")['href'])
            stars = tr.find("span",class_=" staticStars")
            rate =len(stars.find_all("span", class_="staticStar p10"))
            ratings.append(rate)
            users.append(i)
        
        next_page = soup.find("a", {"class":"next_page", "rel":"next"})
        if next_page is not None:
            next_url = next_page.attrs['href']
            url = "https://www.goodreads.com/"+next_url
        else:
            url = None
            
       
    i +=1

    
```

```python
d = {'user':users, 'ratings':ratings,'url':book_urls}
df = pd.DataFrame(data=d)
pd.DataFrame(df).to_csv('bookFile.csv')
```

### Scraper for collecting book features

```python
# Read the bookFile.csv to get all the books we need to scrape
df = pd.read_csv('bookFile.csv',encoding='latin')
book_urls = df['book_urls']
book_urls = pd.Series(book_urls.unique())
```

From user rating data, we have only truncated URLs. We need to complete these URL to correct format, so that we can use them for scrapping book features

```python
def goodreads(url):
    return 'https://www.goodreads.com'+url
book_urls = list(book_urls.apply(goodreads))
```

```python
# this is the main function to scrap the book features from a given book display page url of goodreads
def scrap_bookfeatures(url):

    users=['Googlebot/2.1','Mozilla/5.0']
    agent=random.choice(users)
    req = Request(url, headers={'User-Agent': agent})
    webpage = urlopen(req).read().decode('utf-8')
        
    soup = BeautifulSoup(webpage)
    
    print(url)
    
    bookname = np.nan
    
    if soup.find('h1', {'id': 'bookTitle'}) is not None:
        bookname = soup.find('h1', {'id': 'bookTitle'}).text.strip()
        
        regex = re.compile('\\n+\s+')
        bookname = regex.sub(" ",bookname)
    
    
    avgRating = np.nan
    
    if soup.find('span',{'class': 'average'}) is not None:
        avgRating = soup.find('span',{'class': 'average'}).text
    
    author = np.nan
    
    if soup.find('div', {'class': 'authorName__container'}) is not None:
        author = soup.find('div', {'class': 'authorName__container'}).find('span', {'itemprop': 'name'}).text
    
    
    genre1 = np.nan
    genre2 = np.nan
    
    if soup.find_all('div', {'class': 'elementList '}) is not None:
        genresDiv = soup.find_all('div', {'class': 'elementList '})
        i = 0
        genre = []
        for div in genresDiv:
            if i<2:
                genre.append(div.find('a', {'class': 'actionLinkLite bookPageGenreLink'}).text)
                i += 1
        
        if len(genre) >= 2:
            genre1 = genre[0]
            genre2 = genre[1]
    
    idregex = re.compile('\d+')
    bookid = idregex.search(url)[0]
    
    rating_count = np.nan
    
    if soup.find('span', {'class': 'votes value-title'}) is not None:
        rating_count = soup.find('span', {'class': 'votes value-title'}).text.strip().replace(',','')
    
    language = np.nan
    
    if soup.find('div', {'itemprop': 'inLanguage'}) is not None:
        language = soup.find('div', {'itemprop': 'inLanguage'}).text.strip()
    
    isbn = np.nan
    if soup.find('span', {'itemprop': 'isbn'}) is not None:
        isbn = soup.find('span', {'itemprop': 'isbn'}).text.strip()
    
    rating_count_1 = np.nan
    rating_count_2 = np.nan
    rating_count_3 = np.nan
    rating_count_4 = np.nan
    rating_count_5 = np.nan

    if soup.find('span', {'class': 'rating_graph'}) is not None:
        rating_graph = soup.find('span', {'class': 'rating_graph'}).find('script').text.strip()
        regex = re.compile('\d+')
        rating_graph = regex.findall(rating_graph)
        rating_count_1 = rating_graph[4]
        rating_count_2 = rating_graph[3]
        rating_count_3 = rating_graph[2]
        rating_count_4 = rating_graph[1]
        rating_count_5 = rating_graph[0]
    
    book = {}
    book['url'] = url
    book['bookid'] = bookid
    book['name'] = bookname
    book['avgRating'] = avgRating
    book['isbn'] = isbn
    book['author'] = author
    book['genre1'] = genre1
    book['genre2'] = genre2
    book['lang'] = language
    book['rating_count'] = rating_count
    book['rating1'] = rating_count_1
    book['rating2'] = rating_count_2
    book['rating3'] = rating_count_3
    book['rating4'] = rating_count_4
    book['rating5'] = rating_count_5

    return book
```

```python
# create a list of dictionaries of books
book_dict_list = []
for url in book_urls:
    
    book_dict_list.append(scrap_bookfeatures(url))
    time.sleep(1)
```

```python
# write the book fetures dataframe as a csv
pd.DataFrame(book_dict_list).to_csv('bookfeatures4.csv')
```

## Data Cleaning and Feature Engineering

Our data contains some null values which need to be removed as they may lead to miscalculations of cosine similarity.

Many of our features such as language and genre are categorical. For these we are doing one hot encoding. For other string features such as author, book title, we are using count vectorizer to allow the cosine similarity calculations. 

For authors, we have concatenated first and last names of the author with '-'. This will help us distinguish the authors with same first or last names. For example, if we have author1 "David Macaulay" and author2 "David Shannon", we don't consider them same. Therefore using "David-Macaulay" and "David-Shannon" will be better. 

On the other hand, in book titles, if two books have similar or same keywords, it will mean that two books are very correalted. For example, "Harry Potter Part1" and "Harry Potter Part2" have same keywords, and thus we can say they are correlated. However, we have removed stopwords and punctuations for better results.

In ratings count, we are taking proportions by dividing the values by total rating counts of that particular book. 

We are droping features like ISBN and book URLs as these values are exclusive to the book and will be reduntant in cosine similarity calculations.

### Data Cleaning and Feature Engineering for book data

```python
# Read the book features from csv file
data = pd.read_csv('bookFeatures.csv',encoding='latin')
data.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
## DROP ISBN, Url. These features are unique identifiers to a book and will not be used to find correlations
data = data.drop('isbn',axis=1)
data = data.drop('url',axis=1)

# For missing languages, we are assuming the language is English
data['lang'] = data.lang.replace(np.NaN, 'English')

# Now let us drop any book for which we have missing features. We cannot generalize any other column
data = data.dropna(axis = 0, how="any")

# Now let us normalize the ratings count for a book. We have total ratings available from the scrapped data
data['rating1'] = data['rating1']/data['rating_count']
data['rating2'] = data['rating2']/data['rating_count']
data['rating3'] = data['rating3']/data['rating_count']
data['rating4'] = data['rating4']/data['rating_count']
data['rating5'] = data['rating5']/data['rating_count']

data.head()
```

### Now we need to convert the string values to numerical form, as strings cannot be used in the calculation of cosine similarity

```python
# Converting genre to categorical features
data['genre1'] = data.genre1.str.replace(" ", "")
data['genre2'] = data.genre2.str.replace(" ", '')
data['temp'] = data['genre1'] +' '+ data['genre2']
data.drop('genre1', axis = 1, inplace = True)
data.drop('genre2', axis = 1,  inplace = True)
vect = CountVectorizer()
vect.fit(data['temp'])
cols = vect.transform(data['temp'])
cols = pd.DataFrame(cols.toarray(), columns=vect.get_feature_names())


data = data.reset_index(drop=True)
cols = cols.reset_index(drop=True)
data = pd.concat([data,cols], axis=1, join='inner')
data.drop('temp', axis = 1, inplace = True)

# Converting language to categorical features
df_lang = pd.get_dummies(data['lang'])
data = pd.concat([data, df_lang], axis=1)
data.drop('lang', axis = 1, inplace = True)
```

```python
# Remove stopwords, non-ascii character and punctuations from book name and author and perform a count vectorizer

lemmatizer=nltk.stem.wordnet.WordNetLemmatizer()
stopwords=nltk.corpus.stopwords.words('english')

data.author = data.author.str.replace(" ","-")

#Count Vectorizer object
cv_model = CountVectorizer(stop_words=stopwords, analyzer="word", strip_accents='ascii')

# Count Vectorizer for authors
cv_model.fit(data.author)
cols_author = cv_model.transform(data['author'])
cols_author = pd.DataFrame(cols_author.toarray(), columns=cv_model.get_feature_names())

# Count Vectorizer for book name
regex_string = re.compile(r'[!"#$%&()*+,-./:;<=>?@[\]^_`{|}~\d+]') # Remove punctuations from book title
data.name = data.name.str.replace(regex_string,"")
data.name = data.name.str.lower()
cv_model.fit(data.name)
cols_names = cv_model.transform(data['name'])
cols_names = pd.DataFrame(cols_names.toarray(), columns=cv_model.get_feature_names())

data = pd.concat([data,cols_names], axis=1, join='inner')
data = pd.concat([data,cols_author], axis=1, join='inner')

data.drop('author', axis = 1, inplace =True)
data.drop('name', axis = 1, inplace =True)
```

## Item Based Recommendation Algorithm

### Explain the algorithm

In this algorithm, we are building a matrix of cosine similarities of all books in the catalog by finding books that users tend to read together. The matrix is a symmetric item and item matrix. 

The steps of the algorithm are as follows:

1) For each book we will look for the users who have reviewed that book.

2) Then we will see all the books that those users have reviewed with rating more than 3, this value has been chosen after studying the good reads website, wherein popular books almost always have ratings above three. We are using this rating, as we assume that users will not like to read the book that they have rated low. 

3) Now for all of these books, we will calculate the cosine similarity and add the value to matrix at the respective index positions. If we find that multiple users have bought same set of books along with the first book, we add the cosine values to the existing value, as we assume that more users will like these books together. 

To put this into context, our matrix of cosine similarities is of $m X m$ dimensions.

\begin{equation} \left [ \begin{array}{cccc}
w_{11} & w_{12} & w_{13}  \\
w_{21} & w_{22} & w_{23}  \\
w_{31} & w_{32}  & w_{33}  
\end{array} \right ] \end{equation}


the cosine similarity formula :

For book $i$ and book $k$ having $n$ features the cosine similarity can be calculated as follows:

$W_{ik} = \frac{\sum_{j=1}^n X_{ij}X_{kj}}{\sqrt{\sum_{j=1}^n X_{ij}^2} \sqrt{\sum_{j=1}^n X_{kj}^2}}$

Cosine values range from $[0,1]$ but in our matrix the range will exceed $1$ as we are adding the cosine similarities for every common user.


```python
# read the rating and book features from the csv file
ratings = pd.read_csv('ratings.csv').drop('Unnamed: 0', axis=1)
data_features = pd.read_csv("data_features").drop('Unnamed: 0',axis=1)
```

```python
# all the unique books
books = data_features['bookid']

# Perform innner join to get only the intersection of books and ratings
books = pd.DataFrame(books)
ratings_new = pd.merge(ratings, books, left_on = 'bookid', right_on = 'bookid', how='inner')
```

```python
# Total number of unique books
len(ratings_new['bookid'].unique())
```

We are creating two dictionaries to convert the book id to respective index in matrix and vice versa.

```python
book_id_index = {}
book_index_id = {}
book_unique = sorted(ratings_new['bookid'].unique())

for index in range(len(book_unique)):
    book_id_index[book_unique[index]] = index
    book_index_id[index] = book_unique[index]
    
    
```

Now let us create a inital matrix of n,n zeroes, where n is the number of unique books in the catalog



```python
n = len(book_unique)
matrix = np.zeros((n,n))
```

```python
#Calculation of the main matrix. This is the main part of our algorithm. 
#This matrix calcualtion need to be done only once, and took around 90 minutes for us to run. 
#Since our algorithm is reusing this matrix, it makes our recommendation model scalable. 

row=0

while (row<n):
    book1 = np.array(data_features[data_features['bookid'] == book_index_id[row]].drop('bookid',axis=1))
    # convert to pd Series
    C = list(ratings_new[ratings_new['bookid'] == book_index_id[row]]['user'])
    for c in C:
        book_read_byc1 = list(ratings_new[ratings_new['user'] == c]['bookid'])
        user_ratings = list(ratings_new[ratings_new['user'] == c]['ratings'])
        book_read_byc = [book_read_byc1[i] for i in range(len(book_read_byc1)) if user_ratings[i] >= 3]
        for book2 in book_read_byc:
            # PD SERIES OF book2
            #cosine similarity of book2 and book1
            book2_vector = np.array(data_features[data_features['bookid'] == book2].drop('bookid',axis=1))
            #print(row,book_id_index[book2], matrix[row,book_id_index[book2]])
            matrix[row,book_id_index[book2]] += cosine_similarity(book1, book2_vector)
            if row==book_id_index[book2]:
                matrix[row,book_id_index[book2]] = 0
    print(row)
    row += 1
```

```python
Matrix_DF = pd.DataFrame(matrix)
Matrix_DF.to_csv("Recommendation_Matrix")
```

```python
# As we don't want to calculate matrix everytime, read from csv when kernel restarts
Matrix_DF = pd.read_csv("Recommendation_Matrix")
matrix = np.array(Matrix_DF)
```

The main goal of our project was to give meaningful recommendations to a user based on the books added in the cart. We have achieved this by utilizing database of user ratings on goodreads. We observed from a few iterations of the algorithm that the books recommended match are very meaningful as we can relate the category/genre of the books in cart and the recommended books to be very close.

The core of our algorithm, which is matrix calculation is quite computationally expensive with time complexity of order $ O(m*n*m) $ where m is the number of items and n is the number of users in our database. This is not a major concern as of now, since the matrix calculation is done offline.

But, there is always scope for improvement and we in the next section have listed down some of the limitations we observed and suggest a few improvements.






### Limitations and Further improvements
This is not in real time.

1) If the users have very diverse taste, then the recommendations can be vague.

2) We are considering all the users who have reviewed the book added in the cart. But, it is possible that the user has rated this book low, we would be benefitted if we can discount such users.

3) In this algorithm, we are adding the cosine similarities if two books are matched for multiple users. But the algorithm can be certainly made better by using the user rating as weights.
